## <div style="color: #CC3366;"> 3. Data Analysis

### 3.a Creation of a Sub DataFrame
Here we will now focus on data about Burkina Faso and Ghana. So we will Build a new sub dataFrame which contain data about Burkina Faso and Ghana.

```python
df1 = df[df['Country Name'].isin(['Burkina Faso', 'Ghana'])].copy()
df1
```

### 3.b Statistical Analysis
#### **Measurement of Global Population Growth (1960-2023)**

The objective here is to compute the growth speed of each country


```python
# Burkina Faso Population
bf_pop_2023 = df1.loc[df1['Country Name'] == 'Burkina Faso', '2023'].values[0]
bf_pop_1960 = df1.loc[df1['Country Name'] == 'Burkina Faso', '1960'].values[0]

# Ghana Population
ghana_pop_2023 = df1.loc[df1['Country Name'] == 'Ghana', '2023'].values[0]
ghana_pop_1960 = df1.loc[df1['Country Name'] == 'Ghana', '1960'].values[0]
```

```python
print("====BURKINA FASO====")
bf_growth = ((bf_pop_2023 - bf_pop_1960 )/bf_pop_1960) * 100
print(f"Burkina Faso Growth Speed: {bf_growth}%")

print("====GHANA====")
ghana_growth = ((ghana_pop_2023 - ghana_pop_1960 )/ghana_pop_1960) * 100
print(f"Ghana Growth Speed: {ghana_growth}%")
```

##### **Result Interpretation**

From 1960 to 2024, Burkina Faso and Ghana experienced rapid population growth.

It is also noted that Ghana's population is slightly higher than Burkina Faso's. 

#### **Stability and Dispersion Analysis (Mean & Standard Deviation)**

**The Mean:** This allows for a comparison of the "average size" of the two nations.

**Standard Deviation:** This is the indicator of stability.
A low standard deviation would mean that the population has remained close to its mean (slow/stable growth).
A high standard deviation (as in your case) shows that the data has deviated significantly from the initial mean, which characterizes explosive growth.

```python
years = [col for col in df1.columns if col.isdigit()]
df1['Historical_Mean'] = np.mean(df1[years], axis=1)

df1['Standard_Deviation'] = np.std(df1[years], axis=1)
print(df1[['Historical_Mean', 'Standard_Deviation']])
```

##### **Result Interpretation** 

Burkina Faso (19): Approximately 11.23 million inhabitants on average.
Ghana (83): Approximately 17.82 million inhabitants on average.
What this says: In the long term, Ghana has always had a larger population than Burkina Faso.

Burkina Faso: 5.6 million.
Ghana: 8.18 million.
What this says: The higher this number, the stronger and faster the growth. If the population had remained unchanged since 1960, the standard deviation would be 0.

#### **Critical Threshold Analysis (Tipping Points)**

The goal here is to find the exact year when a country's population reached twice (2x) its 1960 size.

```python
# 1. Calculate the threshold (double the population of 1960) for each country
threshold = df1['1960'] * 2

# 2. Compare the entire years range to the threshold (creates a True/False table)
# .ge() stands for "Greater than or Equal to"
# axis=0 ensures the comparison is done row by row against the threshold
exceeded = df1[years].ge(threshold, axis=0)

# 3. Retrieve the first column name (the year) where the value is True
# idxmax returns the index of the first maximum value (True is > False)
df1['Doubling_Year'] = exceeded.idxmax(axis=1)

# Display the results
print(df1[['1960', 'Doubling_Year']])
```

## <div style="color: #CC3366;"> 4. Data Visualization with Matplotlib

### 4.a Comparative Population Trends (Line Plot)


```python
df1[years].T.plot(figsize=(12, 6), linewidth=2)

plt.title('Population Growth Comparison: Burkina Faso vs Ghana (1960-2023)')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
# Selecting key years
milestones = ['1960', '1990', '2023']
data_milestones = df1[milestones].T # Transpose to have years on X-axis

# Create the bar chart
data_milestones.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'orange'])

# Adding details
plt.title('Population at Key Milestones: 1960, 1990, 2023', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=0) # Keep year labels horizontal
plt.legend(title='Country')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
